Visualize how popularity of Genres has changed over the years. From the graph
one should be able to see for any given year, movies of which genre got released
the most.

```python
genre=pd.read_csv('/content/u.genre.csv',delimiter='|',header=None,names=['genre','index'])
genre
```

```python
genre=genre[genre['genre']!='unknown']
genre
```

```python
merging=pd.merge(left = item ,right= rating, left_on = 'movie id ',right_on ='movieId',how ='inner')
merging.head()
```

```python
final_merging=pd.merge(left = merging ,right= data1, left_on = 'userId',right_on ='user id ',how ='inner')
final_merging.head()
```

```python
final_merging.columns
```

```python
final_merging1=pd.merge(left=genre ,right=final_merging ,left_on='index',right_on='movie id ',how='inner')
```

```python
# year = pd.DatetimeIndex(final_merging[' release date ']).year
# year
final_merging['year']= pd.DatetimeIndex(final_merging[' release date ']).year
final_merging.head()
final_merging.shape
```

```python
sns.lineplot(data = final_merging1 , x = 'genre' , y = ' release date ')
plt.figure(figsize=(18,4))
```
