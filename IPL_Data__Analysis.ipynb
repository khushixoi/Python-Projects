Now , We will merge both the DataFrame

```python
matches=pd.merge(left=IPL_ball_by_ball,
               right=IPL_match,
               left_on='ID',
               right_on ='ID',
               how='left'
               )
```

Now , Data Cleaning

```python
#here we are cheacking if ids are null  or not
matches.isnull().sum()
#as it show 0 , so there are no null id
```

```python
duplicate_id=matches[matches['ID'].duplicated()]
duplicate_id
```

1. From overs 0-19, which over has the most runs scored in, on average?

```python
#here we groupby overs and total_run and  find mean of every over
matches[matches['overs']<=19].groupby("overs")['total_run'].mean().idxmax()
```

2. Top 10 umpires, who were part of the most matches

```python
#In matches Umprire1 list append Umpire2 list and value_counts and take 10 outcomes of umpire1 by using 10 outcome
matches['Umpire1'].append(matches['Umpire2']).value_counts().head(10)
```

3. Which team has the highest probability of winning a match given that they had won the toss?

```python
#in this matches DataFrame we use groupby by winning team and toss winner and value_counts and use idxmax for getting highest probability of winning a match
matches.groupby(['WinningTeam'])['TossWinner'].value_counts().idxmax()
```


4. Top 10 most successful bowlers in terms of number of wickets taken


```python
#in this matches DataFrame we use groupby by isWicketDelivery which is equal to 1 and bowler and value_counts of each bowler and use head for 10 outcomes
matches[matches["isWicketDelivery"]==1]['bowler'].value_counts().head(10)
```

5. Calculate for each team the number of playoff matches player.

```python
# # Filter for playoff match numbers
playoff_matches = matches[matches['MatchNumber'].isin(['Final', 'Qualifier 2','Qualifier 1'])]
# Concatenate 'Team1' and 'Team2' to get a list of all teams in playoff matches
playoff_teams = pd.concat([playoff_matches['Team1'], playoff_matches['Team2']])
# Count the occurrences of each team in playoff matches
team_counts = playoff_teams.value_counts().reset_index()
team_counts.columns = ['Team', 'Playoff_Match_Count']
# Display the team counts in playoff matches
print(team_counts)
```

6. On which venue, the team batting second has the most advantage?

```python
#in this matches dataframe of team2 is equal to WinningTeam and venue use value_counts and use idxmax
matches[matches['Team2']==matches['WinningTeam']]['Venue'].value_counts().idxmax()
```

7. Which batsman in the history of IPL has been caught out the most?

```python
#in this matches in kind columns which is equal to caught  and batter groupby use value_counts and idxmax
matches[matches['kind']=='caught']['batter'].value_counts().idxmax()
```

8. Which bowler has taken the wicket of the batsman in que no. 7?

```python
#in the question 7 we got a batter is 'RG Sharma' and bowler groupby and use value_counts and idxmax for
matches[(matches['kind']=='caught')&(matches['batter']=='RG Sharma')]['bowler'].value_counts().idxmax()
```
