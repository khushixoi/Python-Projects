1. What is the distribution of passengers' ages on the Titanic?


```python
# Create a histogram to visualize the age distribution
sns.histplot(data['Age'])  # Adjust the number of bins as needed
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title(' Distribution of age ')
```

2. How many passengers survived vs. not survived? Visualize the count using a bar plot.

```python
sns.countplot(data=data, x='Survived')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.title('Passenger Survival on the Titanic')
plt.xticks(ticks=[0,1],labels=['Survived','Not Survived'])
```

3. What is the distribution of passenger class (Pclass) on the Titanic?


```python
sns.countplot(data=data,x='Pclass')
plt.title('Distribution of passenger class (Pclass) on the Titanic')
plt.show()
```

4. Visualize the distribution of passenger ages for each passenger class using a box plot.

```python
sns.catplot(x="Pclass", y="Age",kind="box",data=data,orient='vertical');
plt.title('Distribution of passenger ages for each passenger class')
```

5. Create a pair plot to explore relationships between numerical variables like age, fare, and the number of siblings/spouses (sibsp) aboard.


```python
data1=data[['Age','Fare','SibSp']]
sns.pairplot(data1)
```

6. How does the survival rate differ based on passenger class? Visualize it using a bar plot.


```python
# data['Survived'].value_counts()
survival_rate_by_class = data.groupby('Pclass')['Survived'].mean().reset_index()
survival_rate_by_class
sns.barplot(data=survival_rate_by_class, x='Pclass' ,y='Survived')
plt.title('Survival Rate by Passenger Class')
```

7. Create a heatmap to visualize the correlation between numerical features (age, fare, sibsp, parch) in the Titanic dataset.

```python
correlation_matrix = data[['Age', 'Fare', 'SibSp', 'Parch']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='viridis')
plt.title('Correlation between numerical features ')
plt.show()
```

8. How does the survival rate vary based on gender (sex)? Visualize it using a bar plot.

```python
sns.barplot(data=data, x='Sex', y='Survived',ci=None)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender (Sex)')
plt.ylabel('Survival Rate')
plt.show()
```

9. What is the distribution of fares paid by passengers on the Titanic?

```python
distribution_of_fares=data.groupby('Pclass')['Fare'].sum()
sns.histplot(data=data, x='Fare', bins=30, kde=True, color='blue')
plt.title('Distribution of fares paid by passengers')
```

10. Create a violin plot to compare the distribution of ages for survivors and non-survivors.

```python
sns.violinplot(data=data ,x= 'Survived' , y='Age')
plt.title('Distribution of ages for survivors and non-survivors')
```

11. How does the survival rate differ based on the port of embarkation (embarked)? Visualize it using a bar plot.

```python
survival_rate_by_embarked = data.groupby('Embarked')['Survived'].mean().reset_index()
survival_rate_by_embarked
sns.barplot(data=survival_rate_by_embarked, x= 'Embarked' , y='Survived' , ci=None)
plt.title('Survival Rate by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.show()
```

12. Create a scatter plot to explore the relationship between age and fare paid by passengers.


```python
sns.scatterplot(data=data , x='Age', y ='Fare',hue='Survived')
plt.title('Relationship between age and fare paid by passengers')
```
